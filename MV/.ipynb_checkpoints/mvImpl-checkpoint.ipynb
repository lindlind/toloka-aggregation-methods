```python
import numpy as np
import pandas as pd
```

```python
PATH2 = './../TlkAgg2/'
PATH5 = './../TlkAgg5/'
PATH = PATH5
```

```python
data = pd.read_csv(PATH + 'crowd_labels.tsv', sep='\t', header=None)
data.columns = ['worker_id', 'task_id', 'verdict']
data
```

```python
res = data.groupby(['task_id','verdict'])['worker_id'].count()
res = res.unstack(level = 1).fillna(0)
res
```

```python
res['sum'] = res.sum(axis = 1)
res = res.apply(lambda x : x/res['sum'])
del res['sum']
res
```

```python
res['probs'] = res.apply(lambda x : [x.values.tolist()], axis = 1)
if PATH == PATH2:
    res = res['probs'].apply(np.argmax).to_frame()
else:
    res = res['probs'].apply(lambda x : 1 + np.argmax(x)).to_frame()
res
```

```python
res.groupby('task_id').count()['probs'].sum()
```

```python
res.to_csv(PATH + 'mv.tsv', sep = '\t', header=None)
```
